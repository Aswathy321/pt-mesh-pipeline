```python
# importing pandas libraries for developing dataframe
import lxml.html as lx
import pandas as pd
```

```python
url='https://ppp.gov.ph/project-database/'
data=requests.get(url)
par=lx.fromstring(data.content)
row=par.xpath('//tr')
```

```python
[len(T)for T in row[:23]]
```

```python
col=[]
i=0
for r in row[0]:
    i+=1
    nm=r.text_content()
    print('%d:"%s"'%(i,nm))
    col.append((nm,[]))
```

```python
for j in range(0,len(row)):
    R=row[j]
    if len(R)!=10:
        break
    i=0
    for c in R.iterchildren():
        text=c.text_content()
        if i>0:
            try:
                text=int(text)
            except:
                pass
        col[i][0].append(text)
        i+=1
```

```python
[len(C)for (title,C)in col]
```

```python
#Implementing proposed columns on to a dataframe

dic={title:colm for(title,colm)in col}
df=pd.DataFrame(dict)
df.head()
df.to_csv('scrape_data.csv',header=False,index=False)
```

```python
# Another method for scraping
import urllib.request
from pprint import pprint

from html_table_parser.parser import HTMLTableParser
import pandas as pd

def url_get_contents(url):

    req = urllib.request.Request(url=url)
    f = urllib.request.urlopen(req)
    return f.read()
xhtml = url_get_contents('https://ppp.gov.ph/project-database/').decode('utf-8')
p = HTMLTableParser()
p.feed(xhtml)
pprint(p.tables[1])

print("\n\nPANDAS DATAFRAME\n")
print(pd.DataFrame(p.tables[1]))
```
